# 네이버 "실시간" 검색어(1위~20위)
파이썬 BeautifulSoup을 이용한 웹 크롤러

출처 : https://baobob1024.tistory.com/87?source=post_page-----9d6abf5a7296----------------------

```python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://naver.com") # connection
# naver에서 소스 받아오기
html = req.text

# BeautifulSoup으로 html 소스 -> 객체
# BeautifulSoup(html 소스코드, 사용할 parser타입)
# html.parser는 파이썬 내장 함수
soup = BeautifulSoup(html, 'html.parser')
rtKeyword = soup.select\
    ('div.ah_roll.PM_CL_realtimeKeyword_rolling_base > div > ul > li')

b = []
# 태그 내 문자열을 b 리스트에 추가
for sill in rtKeyword:
    b.append(sill.text)

k = 1;
list_rtKeyword = []
# 헤더 출력
print("="*30 + '\n' + ' '*7 + "NAVER RANK LIST\n" + '='*30)

# 문자열에서 핵심 문자열만 list_rtKeyword에 추가
for i in b:
    if k > 9 :
        list_rtKeyword.append(i[5: -2])
    else :
        list_rtKeyword.append(i[4: -2])
    k += 1

# 실검 출력
# s는 순워, list는 목록 요소 접근 가능
for s, list in enumerate(list_rtKeyword):
    print("%d위" %(s+1) + " " + list)
```

임의의 시간대로 필터링하여 과거 순위를 조회하면,

**https://datalab.naver.com/keyword/realtimeList.naver?datetime=2019-11-18T20%3A33%3A00**

쿼리 스트링 부분에 datetime이 생긴 것을 볼 수가 있음.

요 부분을 조작하여 크롤링하는 것이 키포인트★

```python
# 해당 url로 post 요청
import requests as rq
from bs4 import BeautifulSoup

url = 'https://datalab.naver.com/keyword/realtimeList.naver?datetime=2019-11-24T18%3A39%3A00'

res = rq.get(url)
soup = BeautifulSoup(res.content, 'lxml')
print(soup)

rank_tabs = soup.select('keyoword_rank')[0:4]
print(rank_tabs)
```
